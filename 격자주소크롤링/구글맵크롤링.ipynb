```python
from selenium.webdriver import Chrome
import time
import os
import pandas as pd
from selenium import webdriver
#팝업창
from selenium.webdriver.common.alert import Alert
```

```python
#옵션설정
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
```

```python
driver = webdriver.Chrome('D:/chromedriver_win32/chromedriver.exe', options=options)
driver.implicitly_wait(3) #웹페이지 전체가 넘어올때까지 기다리기
```

```python
#url 접근
driver.get('https://www.google.com/maps/@37.5555764,127.0078127,14z?hl=ko')
time.sleep(5)
```

```python
location=pd.read_csv('우선취약법정동+위경도.csv', encoding = 'utf-8-sig')
location
```

```python
location['주소']=str(0)
location
```

```python
text = []

for i in range(13042,location.shape[0]):
    y=location.iloc[i,18]
    x=location.iloc[i,17]
    xy=str(y)+','+' '+str(x)
    
    #위경도검색
    driver.find_element_by_xpath('//*[@id="searchboxinput"]').send_keys(xy) 
    driver.find_element_by_xpath('//*[@id="searchbox-searchbutton"]').click()
    time.sleep(3)
    
    #주소가져오기
    a=driver.find_element_by_xpath('//*[@id="pane"]/div/div[1]/div/div/div[7]/div/div[1]/span[3]/span[3]') 
    
    
    text.append(a.text)
    num=' '+str(i)+'번'
    print(f'{a.text}\n {num}진행 중!')
    
    #검색어 지우기
    driver.find_element_by_xpath('//*[@id="searchbox"]/a').click()
    
```
